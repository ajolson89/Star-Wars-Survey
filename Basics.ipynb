# Import Data and Look at First Rows

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.columns
```

# Clean Up Data

```python
yes_no = {
    'Yes': True,
    'No': False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'])
```

```python
print(star_wars['Unnamed: 4'], star_wars['Unnamed: 5'], star_wars['Unnamed: 6'], star_wars['Unnamed: 7'], star_wars['Unnamed: 8'])

print(star_wars.columns[3])
```

# Map Data Correctly

```python
import numpy as np
episode_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}


for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(episode_map)

star_wars = star_wars.rename(columns={    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
     })                                 
        

    
star_wars.head()
```

```python
print(star_wars[star_wars.columns[9]])
```

```python
star_wars = star_wars.rename(columns={  
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "rank_1",
    "Unnamed: 10": "rank_2",
    "Unnamed: 11": "rank_3",
    "Unnamed: 12": "rank_4",
    "Unnamed: 13": "rank_5",
    "Unnamed: 14": "rank_6"
     })      

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.head()
```

# Plots Comparing the 6 Films

```python
%matplotlib inline
import matplotlib.pyplot as plt
means = star_wars[star_wars.columns[9:15]].mean()
labels = ['The Phantom Menace', 'Attack of the Clones', 'Revenge of the Sith', 'A New Hope', 'The Empire STrikes Back', 'Return of the Jedi']
ra = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5]

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
plt.xticks(ra, labels, rotation='vertical')
plt.title("Star Wars Rank by Movie")
plt.show()
```

```python
sums = star_wars[star_wars.columns[3:9]].sum()

plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
plt.xticks(ra, labels, rotation='vertical')
plt.title("Respondents Seen Star Wars Movie")
plt.show()
```

# Movie Rank by Gender Barplot

```python
# Establish Male/Female Database
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
means = males[males.columns[9:15]].mean()
meansf = females[females.columns[9:15]].mean()

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
locs = np.arange(len(means))
offset_locs = locs + 0.35

bar_1 = ax.bar(left = locs, height = means, width = 0.35)
bar_2 = ax.bar(left = offset_locs, height = meansf, width = 0.35, color = "green")


plt.xticks(ra, labels, rotation='vertical')
plt.title("Movie Rank by Gender")
plt.legend((bar_1, bar_2), ("Male Rating", "Female Rating"), loc="upper right")
plt.show()
```

In general, movie rank followed the same trend across genders.
